```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0, 5 ,11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional
plt.plot(x, y, color = 'red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols = 2)
```

```python
for axe in ax: 
    axe.plot(x, y, 'b')
    axe.set_title('Titulo')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo')

ax[1].plot(x, y, 'b')
ax[1].set_title('Titulo')

fig
```

```python
fig, ax = plt.subplots(nrows = 5, ncols = 5)
plt.tight_layout() # Organiza o layout
```

```python
fig = plt.figure(figsize=(8, 4), dpi = 100)
```

```python
fig, axes = plt.subplots(figsize = (12, 3))

axes.plot(x, y, 'r-', label= 'x ^ 2')
axes.plot(y, x, 'g--', label= 'x ** 1/2')
axes.set_title('Título')
axes.legend(loc = 0)
```
